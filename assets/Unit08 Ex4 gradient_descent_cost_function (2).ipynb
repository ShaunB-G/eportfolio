## Calculating cost with gradient descent and learning rate
- Change the iteration and learning rate vaules and see the impact on cost. 
- Low iteration values with high learning rate (i.e. big steps) may lead to miss the global minimum
- Goal is to reach minimum cost with minimum iteration

```python
# code credit:codebasics https://codebasics.io/coming-soon

import numpy as np

def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 100       #change value
    n = len(x)
    learning_rate = 0.08   #change value

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print ("m {}, b {}, cost {} iteration {}".format(m_curr,b_curr,cost, i))

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```

```python
import numpy as np


#Altered function to include iterations and learning_rate as an input
def gradient_descent(x, y, iterations, learning_rate):
    # Initialization of slope (m) and intercept (b)
    m_curr = b_curr = 0
    
    # Number of data points
    n = len(x)

    # Gradient Descent Iterations
    for i in range(iterations):
        # Predicted values based on current parameters
        y_predicted = m_curr * x + b_curr
        
        # Cost calculation using Mean Squared Error
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])
        
        # Partial derivatives of the cost function with respect to parameters
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        
        # Parameter updates using the gradient and learning rate
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
        # Display current parameters, cost, and iteration
        print("m {}, b {}, cost {} iteration {}".format(m_curr, b_curr, cost, i))

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

```

```python
# Experiment 1: Default values
print("\nExperiment 1: Default values (iterations=100, learning_rate=0.08)")
gradient_descent(x, y, iterations=100, learning_rate=0.08)
 
```

```python
# Experiment 2: Higher Iterations
print("\nExperiment 2: Higher iterations (iterations=500, learning_rate=0.08)")
gradient_descent(x, y, iterations=500, learning_rate=0.08)
```

Increasing Iterations: 
Increasing the iterations saw a vast improvement in the reduction of the cost value
With this data did not noticabily increase the processing time. However, it might lead to longer computation times for larger datasets

Generally helps in achieving a more accurate and refined convergence. However, it might lead to longer computation times and diminishing returns beyond a certain point.



```python
# Experiment 3: Lower Iterations
print("\nExperiment 3: Higher iterations (iterations=50, learning_rate=0.08)")
gradient_descent(x, y, iterations=50, learning_rate=0.08)
```

Decreasing Iterations: 
Decreasing the iterations saw a vast increase of the cost value by the last iteration.

Having already completed a experiment with more iterations it is obvious that this number of iterations is no where near enough and increasing the iterations will vastly reduce the cost without any negative outcomes.

```python
# Experiment 4: Higher Learning Rate
print("\nExperiment 4: Higher learning rate (iterations=100, learning_rate=0.2)")
gradient_descent(x, y, iterations=100, learning_rate=0.2)
```

Increasing Learning Rate: 
A small increase in the learning rate caused the cost to increase with each itteration this suggests that the minimal was missed 
Can lead to faster convergence but risks overshooting the minimum/oscillating around it. Choosing a learning rate that is too high may prevent convergence altogether.

```python
# Experiment 5: Lower Learning Rate
print("\nExperiment 5: Lower learning rate (iterations=100, learning_rate=0.01)")
gradient_descent(x, y, iterations=100, learning_rate=0.01)
```

Decreasing Learning Rate: 
The algorithm required more incriments to reduce the cost. Slowing down the convergence can provide stability, especially in cases where a higher learning rate would cause overshooting. However, it might require significantly more iterations to find the minimal.

Key takeaways:
Finding the Right Balance: The optimal combination of iterations and learning rate depends on the specific characteristics of the dataset and the processing power/time availble. It may involves experimentation and fine-tuning to strike the right balance.

Trade-offs: There is a trade-off between convergence speed and stability. Increasing one may come at the expense of the other, and finding the right compromise is crucial for effective gradient descent.

Monitoring Convergence: Regularly monitoring the cost values and convergence behavior during training helps in identifying the effectiveness of the chosen hyperparameter values and making adjustments as needed.

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y, iterations, learning_rate):
    m_curr = b_curr = 0
    n = len(x)
    costs = []  # List to store the cost values during iterations

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

        # Append the current cost to the list for plotting
        costs.append(cost)

        # Display current parameters, cost, and iteration (optional)
        # print("m {}, b {}, cost {} iteration {}".format(m_curr, b_curr, cost, i))

    # Plotting the cost values during iterations
    plt.plot(range(iterations), costs, label=f'LR={learning_rate}, Iter={iterations}')

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Run experiments with different iteration and learning rate values
plt.figure(figsize=(10, 6))

# Experiment 1: Default values
gradient_descent(x, y, iterations=100, learning_rate=0.08)

# Experiment 4: Lower learning rate
gradient_descent(x, y, iterations=100, learning_rate=0.01)

plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Effect of Iterations and Learning Rate on Cost')
plt.legend()
plt.show()
```
